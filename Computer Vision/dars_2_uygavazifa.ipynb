```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
%matplotlib inline#bu bizga rasmni yangi oynda ochish uchun kerak
```

```python
img = cv2.imread("rasmlar/farq.jpg")
```

```python
print(img.shape[1])
print(img.shape[2])
print(img.shape[0])#rasmning shapelari tekshirdik
```

```python
delta = 3#bizda rasmning ortasida chizziqni bolish uchun kerak
```

```python
img_left = img[:, :img.shape[1]//2-delta,]
img_right = img[:,img.shape[1]//2+delta:,]#rasmni 2 ga bo'lib olamiz
```

```python
print(img_left.shape)
print(img_right.shape)#bo'lingan rasmlarning shapelari
```

```python
difference = 255 - cv2.absdiff(img_left, img_right)#dif funksiyasi orqali 2 ta rasmning farqi kelib chiqadi
difference[difference<30] = 255#bunda 30 dan kichiklarini oq ga aylantrib yuboradi
cv2.imshow('diff', difference)
cv2.waitKey(0)

cv2.destroyAllWindows()
```

```python
fig, (plt1, plt2, plt3, plt4) = plt.subplots(nrows=1, ncols=4, figsize=(12,2))
```

```python
%matplotlib qt#yangi oynada chiqishni bekor qilish
```

```python
#Change color chanel
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
img_left = cv2.cvtColor(img_left, cv2.COLOR_BGR2RGB)
img_right = cv2.cvtColor(img_right, cv2.COLOR_BGR2RGB)
difference = cv2.cvtColor(difference, cv2.COLOR_BGR2RGB)

#plotting 
plt1.set_title("Orginal")
plt1.imshow(img)

plt2.set_title("Left image")
plt2.imshow(img_left)

plt3.set_title("Right image")
plt3.imshow(img_right)

plt4.set_title("Difference images")
plt4.imshow(difference)

fig#bu yerda rasmlarni solishtrib ko'ryapmiz
```
