```python
rasm_opencv = cv2.imread("rasmlar/tashkent.jpg")
rasm_opencv = cv2.cvtColor(rasm_opencv,cv2.COLOR_BGR2RGB)
plt.imshow(rasm_opencv,cmap="gray")
```

# Subplot usullari:

* figure -> subplot
* subplots

```python
figure, (plt1,plt2,plt3) = plt.subplots(1,3,figsize=(9,3))
plt1.set_title("Red")
plt1.set_ylabel("Balandligi")
plt1.set_xlabel("Eni")
plt1.imshow(red_channel)

plt2.set_title("Green")
plt2.set_ylabel("Balandligi")
plt2.set_xlabel("Eni")
plt2.imshow(green_channel)

plt3.set_title("Blue")
plt3.set_ylabel("Balandligi")
plt3.set_xlabel("Eni")
plt3.imshow(blue_channel)
```

```python
plt.figure(figsize=(10,12))

for i in range(1,31):#n(30) ta rasmni ekranga chiqarish uchun
    plt.subplot(6,5,i)#ketma ket joylashtrish uchun
    plt.imshow(np.random.randint(0,255,size=(300,300),dtype=np.uint8))
    
plt.show()
```
